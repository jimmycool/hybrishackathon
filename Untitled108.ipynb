```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import keras as keras
from keras.layers import Input,Conv2D,MaxPooling2D,UpSampling2D,Flatten,Conv2DTranspose,concatenate,BatchNormalization
from keras.models import Model, Sequential
from keras.layers.core import Dense, Dropout
from keras.layers.advanced_activations import LeakyReLU
from keras.datasets import mnist
from keras import initializers
from keras import backend as K
import tensorflow as tf
from keras.utils.vis_utils import plot_model
from tensorflow.keras.losses import binary_crossentropy
act='LeakyReLU'
os.environ["KERAS_BACKEND"] = "tensorflow"
def build_model(input_layer, start_neurons):
    conv1 = Conv2D(start_neurons * 1, (3, 3), activation=act, padding="same")(input_layer)
    conv1=BatchNormalization()(conv1) 
    conv1=Dropout(0.3)(conv1)
    conv1 = Conv2D(start_neurons * 1, (3, 3), activation=act, padding="same")(conv1)
    conv1=BatchNormalization()(conv1) 
    conv1=Dropout(0.3)(conv1)

    pool1 = MaxPooling2D((2, 2))(conv1)
    pool1=BatchNormalization()(pool1) 
    conv2 = Conv2D(start_neurons * 2, (3, 3), activation=act, padding="same")(pool1)
    conv2=BatchNormalization()(conv2)
    conv2=Dropout(0.3)(conv2)
 
    conv2 = Conv2D(start_neurons * 2, (3, 3), activation=act, padding="same")(conv2)
    conv2=BatchNormalization()(conv2) 
    conv2=Dropout(0.3)(conv2)

    pool2 = MaxPooling2D((2, 2))(conv2)
    pool2=BatchNormalization()(pool2) 

    conv3 = Conv2D(start_neurons * 4, (3, 3), activation=act, padding="same")(pool2)
    conv3=BatchNormalization()(conv3) 
    conv3=Dropout(0.3)(conv3)

    conv3 = Conv2D(start_neurons * 4, (3, 3), activation=act, padding="same")(conv3)
    conv3=BatchNormalization()(conv3) 
    conv3=Dropout(0.3)(conv3)

    pool3 = MaxPooling2D((2, 2))(conv3)
    pool3=BatchNormalization()(pool3) 

    conv4 = Conv2D(start_neurons * 8, (3, 3), activation=act, padding="same")(pool3)
    conv4=BatchNormalization()(conv4) 
    conv4=Dropout(0.3)(conv4)

    conv4 = Conv2D(start_neurons * 8, (3, 3), activation=act, padding="same")(conv4)
    conv4=BatchNormalization()(conv4) 
    conv4=Dropout(0.3)(conv4)

    pool4 = MaxPooling2D((2, 2))(conv4)
    pool4=BatchNormalization()(pool4) 

    convm = Conv2D(start_neurons * 16, (3, 3), activation=act, padding="same")(pool4)
    convm=BatchNormalization()(convm) 
    convm=Dropout(0.3)(convm)

    convm = Conv2D(start_neurons * 16, (3, 3), activation=act, padding="same")(convm)
    convm=BatchNormalization()(convm) 
    convm=Dropout(0.3)(convm)

    deconv4 = Conv2DTranspose(start_neurons * 8, (3, 3), strides=(2, 2), padding="same")(convm)
    deconv4=BatchNormalization()(deconv4) 
    deconv4=Dropout(0.3)(deconv4)

    uconv4 = concatenate([deconv4, conv4],axis=-1)
    uconv4=BatchNormalization()(uconv4) 
    uconv4=Dropout(0.3)(uconv4)

    uconv4 = Conv2D(start_neurons * 8, (3, 3), activation=act, padding="same")(uconv4)
    uconv4=BatchNormalization()(uconv4) 
    uconv4=Dropout(0.3)(uconv4)
    uconv4=Dropout(0.3)(uconv4)

    uconv4 = Conv2D(start_neurons * 8, (3, 3), activation="relu", padding="same")(uconv4)
    uconv4=BatchNormalization()(uconv4) 
    uconv4=Dropout(0.3)(uconv4)

    deconv3 = Conv2DTranspose(start_neurons * 4, (3, 3), strides=(2, 2), padding="same")(uconv4)
    deconv3=BatchNormalization()(deconv3) 
    deconv3=Dropout(0.3)(deconv3)

    uconv3 = concatenate([deconv3, conv3],axis=-1)
    uconv3 = Conv2D(start_neurons * 4, (3, 3), activation=act, padding="same")(uconv3)
    uconv3=BatchNormalization()(uconv3) 
    uconv3=Dropout(0.3)(uconv3)

    uconv3 = Conv2D(start_neurons * 4, (3, 3), activation=act, padding="same")(uconv3)
    uconv3=BatchNormalization()(uconv3) 
    uconv3=Dropout(0.3)(uconv3)

    deconv2 = Conv2DTranspose(start_neurons * 2, (3, 3), strides=(2, 2), padding="same")(uconv3)
    deconv2=BatchNormalization()(deconv2) 
    deconv2=Dropout(0.3)(deconv2)

    uconv2 = concatenate([deconv2, conv2],axis=-1)
    uconv2 = Conv2D(start_neurons * 2, (3, 3), activation=act, padding="same")(uconv2)
    uconv2 = Conv2D(start_neurons * 2, (3, 3), activation=act, padding="same")(uconv2)
    uconv2=Dropout(0.3)(uconv2)

    deconv1 = Conv2DTranspose(start_neurons * 1, (3, 3), strides=(2, 2), padding="same")(uconv2)
    uconv1 = concatenate([deconv1, conv1],axis=-1)
    uconv1 = Conv2D(start_neurons * 1, (3, 3), activation=act, padding="same")(uconv1)
    uconv1=Dropout(0.3)(uconv1)

    output_layer = Conv2D(3, (2,2), padding="same",activation='sigmoid')(uconv1)

    model = Model(inputs=input_layer, outputs=output_layer, name="noise_model")
    return model

def dice_coefficient(y_true, y_pred):
    numerator = 2 * tf.reduce_sum(y_true * y_pred)
    denominator = tf.reduce_sum(y_true + y_pred)

    return (numerator / (denominator + tf.keras.backend.epsilon()))
def loss(y_true, y_pred):
    return binary_crossentropy(y_true, y_pred) - K.log(dice_coefficient(y_true, y_pred) + tf.keras.backend.epsilon())
input_layer = Input((256,256,3))
from focal_loss import BinaryFocalLoss
modef = build_model(input_layer,64)
optimizer = tf.keras.optimizers.SGD(learning_rate=0.09,momentum=0.95,nesterov=True)
modef.compile(loss=BinaryFocalLoss(gamma=2), optimizer=optimizer,metrics=['accuracy','Precision','Recall'])
modef.fit(np.array(images1[0:310]),np.array(images_cupseg[0:310]).astype('float32'),epochs=100,batch_size=32)
```
